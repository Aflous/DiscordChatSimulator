Test reading a script

```python
with open('test_script.txt') as f:
    lines = f.read().splitlines()
```

```python
prof_pic_dict = {'Billy': 'profile_pictures/prof.PNG',
                'Shregory': 'profile_pictures/shregory.PNG'}
```

```python
name_up_next = False
current_name = None
current_lines = []
finished_message = False
ims = []
msg_number = 1

for line in lines:
    # If line is empty (signifying end of message chain)
    if line=='':
        name_up_next=True
        current_lines = []
        continue
    # lines that are commented out
    if line[0]=='#':
        continue
    # If currently on a name line
    if name_up_next:
        current_name = line.split(':')[0]
        name_up_next = False
        continue
    # otherwise append message
    current_lines.append(line)
    generate_chat(current_lines, current_name, '9:43', prof_pic_dict[current_name]).save(f'ims/{msg_number:03d}{current_name[0]}.png')
    msg_number+=1
    
```
